```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data_new = pd.read_csv('5-16/MokuDataLoggerData_19700101_103305.csv', skiprows= 7)
data_old = pd.read_csv('4-15/locked/data_0_1744777187.3391218.csv', skiprows= 6, header=None)
print(data_old.head())
```

```python
time_old = data_old[2]
fp_old = data_old[0]
sweep_old = data_old[1]


fig, axes = plt.subplots(2, 1, figsize=(10, 8))



axes[0].set_title('Old Data')
axes[0].scatter(time_old,fp_old*50)
axes[0].plot(time_old,sweep_old)
axes[0].set_xlabel('Time (s)')
axes[0].set_ylabel('FP (V)')
axes[0].set_xlim(-0.0065, 0.0065)



time_new = data_new['% Time (s)']
fp_new = data_new[' Input 1 (V)']
sweep_new = data_new[' Input 2 (V)']

axes[1].set_title('New Data')
axes[1].scatter(time_new,fp_new*50)
axes[1].plot(time_new,sweep_new)
axes[1].set_xlabel('Time (s)')
axes[1].set_ylabel('FP (V)')
axes[1].set_xlim(0.0085, 0.022)

plt.tight_layout()
```

```python
old_timest = time_old[1] - time_old[0]
new_timest = time_new[1] - time_new[0]
print(f"Old time step: {old_timest}")
print(f"New time step: {new_timest}")

old_samp = 1/old_timest
new_samp = 1/new_timest
print(f"Old sample rate: {old_samp}")
print(f"New sample rate: {new_samp}")
```
